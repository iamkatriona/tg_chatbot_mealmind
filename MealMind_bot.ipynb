```python
@bot.callback_query_handler(func=lambda call: call.data in ['breakfast', 'lunch', 'dinner'])

def callback(call):
    x = 3
    
    if call.data == 'breakfast':
        x = 1
    elif call.data == 'lunch':
        x = 2
        
    
    con = sl.connect('database_mm.db', timeout=20)
    cur = con.cursor()
    
    data_user_pref = cur.execute('SELECT u_meat, u_lactose, u_citrus FROM users WHERE u_id = ?', (call.message.chat.id,))
    for i in data_user_pref:
        u_m, u_l, u_c = i 
    
    sql_break = '''SELECT r_key, r_name, products, steps
                    FROM recepies
                    WHERE (r_meat <= ? AND r_lactose <= ?) AND (r_citrus <= ? AND meal_id = ?)
                    ORDER BY RANDOM()
                    LIMIT 1'''
    
    res_give = cur.execute(sql_break, (u_m, u_l, u_c, x))
    res_give = cur.fetchall()
    r_key_for_like = 1
    res_res = ' '
    
    if res_give:
        res_res = '\n\n'.join(res_give[0][1:])
        r_key_for_like = res_give[0][0]
    
    markup = types.InlineKeyboardMarkup(row_width=2)
    item_like = types.InlineKeyboardButton(emojize(":red_heart:", variant="emoji_type"), callback_data=f'like {r_key_for_like}')
    item_next = types.InlineKeyboardButton(emojize(":next_track_button:"), callback_data=call.data)
    markup.add(item_like, item_next)
    
    if res_give:
        bot.send_message(call.message.chat.id, res_res, reply_markup=markup)
    else:
        bot.send_message(call.message.chat.id, 'Нет пока рецептов, удовлетворяющих вашим ограничениям')
    
    con.commit()
    con.close()
     
```

```python
@bot.message_handler(commands = ["favourites"])

def main_func(message):
    con = sl.connect('database_mm.db', timeout=20)
    cur = con.cursor()
    
    sql_fav = '''SELECT DISTINCT recepies.r_name
               FROM recepies
                    INNER JOIN likes ON recepies.r_key = likes.l_r_key
               WHERE likes.l_u_id = ?
               ORDER BY 1'''
    fav_res = cur.execute(sql_fav, (message.chat.id,))
    fav_res = cur.fetchall()

    if fav_res:
        s = ''
        for i in fav_res:
            s+= i[0] + '\n'

        bot.send_message(message.chat.id, s)
    else:
        bot.send_message(message.chat.id, 'Пока в избранном ничего нет')
    
    con.commit()
    con.close()
```
